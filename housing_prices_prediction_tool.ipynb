```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_absolute_error

data = {
    'Square_Feet': [500, 700, 1000, 1200, 1500, 1800, 2000, 2500],
    'Price_in_Lakhs': [25, 38, 55, 62, 78, 88, 105, 130]
}

df = pd.DataFrame(data)

print(df)
```

```python
data = {
    'Square_Feet': [500, 700, 1000, 1200, 1500, 1800, 2000, 2500],
    'Price_in_Lakhs': [25, 38, 55, 62, 78, 88, 105, 130]
}

df = pd.DataFrame(data)

print(df)
```

```python
plt.scatter(df['Square_Feet'], df['Price_in_Lakhs'], color='blue', marker='o')
plt.title('House Size vs Price')
plt.xlabel('Square Footage (sq ft)')
plt.ylabel('Price (in Lakhs)')
plt.grid(True)
plt.show()
```

```python
X = df[['Square_Feet']] # Matrix (Capital X,)
y = df['Price_in_Lakhs'] # Vector (Small y)
```

```python
# 1. setting the ai model for linear regression'
model = LinearRegression()

# 2. training the model using false dataset
model.fit(X, y)

# 3. asking ai to predict prices
predictions = model.predict(X)

# 4. plotting blue dots
plt.scatter(X, y, color='blue', label='Actual Data')

# 5. Plot the AI's red line of best fit
plt.plot(X, predictions, color='red', linewidth=2, label='AI Regression Line')

# 6. presentation uplifting
plt.xlabel('Square Footage')
plt.ylabel('Price (Lakhs)')
plt.legend()
plt.grid(True)
plt.show()
```

```python
# assuming the size of the house required
new_house_size = [[1600]]

# asking the model to predict the price based on data assumed/given as input
predicted_price = model.predict(new_house_size)

print(f"For a house with 1600 sq ft, the AI predicts a price of: {predicted_price[0]:.2f} Lakhs")
```

```python
# Calculate the average error of our red line
mae = mean_absolute_error(y, predictions)

print(f"Model Evaluation Complete!")
print(f"Mean Absolute Error (MAE): {mae:.2f} Lakhs")
```
